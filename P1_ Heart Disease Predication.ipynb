```python
#Downloading the dependencies

import numpy as np 
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
```

```python
heart_data = pd.read_csv("/python-apt/data.csv") #Loading CSv file to a variable
```

```python
heart_data.tail()
```

```python
heart_data.shape #Number of rows and columns
```

```python
heart_data.info()
```

```python
#Statistical Measures of Data

heart_data.describe()
```

```python
#checking the distribution of target variable

heart_data['target'].value_counts()
```

1 --------> DEFECIVE HEART
0 -------->
 HEALTHY HEART


```python
#SPLITTING THE FEATURES AND TARGET
X=heart_data.drop(columns='target', axis=1)
Y=heart_data['target']
```

```python
Y
```

```python
X
```

 **SPLITTING TRAINING AND TESTING DATA**

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=2 )
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
Y_train.shape

```

```python
Y_test.shape
```

**LOGISTIC REGRESSION**

```python

model = LogisticRegression()
```

```python
#training the model with train data
model.fit(X_train,Y_train)
```

```python
#Model evaluation-By using accuracy score as our Model evaluation metric
#predicting training data


X_train_prediction = model.predict(X_train)
training_data_accuracy= accuracy_score(X_train_prediction,Y_train)
```

```python
print('Accuracy on training data:',training_data_accuracy) 
```

```python
#comparing test and train target values
X_test_prediction = model.predict(X_test)
testing_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy on testing data:',testing_data_accuracy) 
```
